```python
import pandas as pd

# Utils
import src.utils.analysis_basics.check_constraints as check
import src.utils.visualization.gantt_diagram as gantt
from src.utils.analysis_basics.check_lateness import get_jobs_with_lateness_metrics, get_jobs_aggregated
from src.utils.editor import enrich_schedule_dframe
from src.utils.rolling_process.rolling_filter import *

# Solver
# from src.solvers.converter import get_schedule_dframe, get_job_ops_dict, get_times_dict
from src.solvers.lp.tardiness import solve_jssp_sum_tardiness_minimization

# Data access
from configs.path_manager import get_path

pd.set_option('display.min_rows', 12)
pd.set_option('display.max_rows', 16)
```

```python
# Hard solver termination condition
max_time = 60*15     # 15 min

# File paths
logs_path = get_path("data", "logs")
basic_data_path = get_path("data", "basic")
```

```python
# Load file
df_routings = pd.read_csv(basic_data_path / "ft10_routings.csv")
df_routings
```

### Load jobs information (15 jobs) and generate the job-shop scheduling problem

```python
df_jobs_times_final = pd.read_csv(basic_data_path / "ft10_jobs_times.csv")
df_jobs_times = df_jobs_times_final.head(15)
df_jobs_times
```

```python
df_jssp = df_jobs_times[["Job", "Routing_ID"]].merge(df_routings, on="Routing_ID",how="left")
df_jssp
```

## Scheduling

#### Preparation

```python
job_ops = get_job_ops_dict(df_jssp)
job_ops
```

```python
times_dict = get_times_dict(df_jobs_times, earliest_start_column="Ready Time")
times_dict
```

### HiGHS with PuLP
> high performance serial and parallel solver


```python
log_file_path = logs_path / "tardiness_highs.log" if logs_path else None

schedule = solve_jssp_sum_tardiness_minimization(
    job_ops=job_ops,
    times_dict=times_dict,
    solver_type="HiGHS",
    msg=False,
    solver_time_limit=max_time,
    solver_relative_gap_limit=0.01,
    log_file=log_file_path
)
```

```python
df_schedule = get_schedule_dframe(schedule)
df_schedule
```

```python
gantt.get_plot(df_schedule, perspective="Machine")
check.check_core_schedule_constraints(df_schedule)
check.is_duration_correct(df_schedule)
```

```python
check.is_start_correct(df_schedule, df_jobs_times)
```

#### Tardiness Check

```python
df_schedule_with_times = enrich_schedule_dframe(df_schedule, df_jobs_times, on= "Job")
df_last_ops_with_metrics = get_jobs_with_lateness_metrics(df_schedule_with_times)
get_jobs_aggregated(df_last_ops_with_metrics, column="Tardiness")
```
