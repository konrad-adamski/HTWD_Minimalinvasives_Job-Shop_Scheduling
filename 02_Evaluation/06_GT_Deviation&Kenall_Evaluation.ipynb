```python
import matplotlib.pyplot as plt
import pandas as pd

# Analyses

df_experiments = pd.read_csv("data/experiments_v01_gt.csv")
df_experiments
```

```python
df_shift_dev = pd.read_csv("data/experiment_shift_dev_v01_gt.csv")

df_dev_mean = (
    df_shift_dev
    .groupby("Experiment_ID")["Deviation",]
    .mean()
    .round(2)  # ➜ Rundet auf 2 Nachkommastellen
    .reset_index()
    .rename(columns={"Deviation": "Deviation_Mean", "Pairs": "Pairs_Mean"})
)
df_experiments = df_experiments.merge(df_dev_mean, on="Experiment_ID", how="left")
df_experiments
```

## Evaluation

```python
from src.DataFrameEvaluations import plot_experiment_boxrow, plot_experiment_lines_compare
```

```python
fig, axes = plot_experiment_boxrow(
    df_values=df_shift_dev,
    df_meta=df_experiments,
    value_col="Deviation",
    x_col="Experiment_Type", x_col_as= "",
    # optional Facetten, falls gewünscht:
    #col_col="Sim Sigma",
    value_as="",
    median_fmt = ".1f",
    ratio_label_on=False,
    ymax= 1300
)
fig.savefig("output/Evaluation_GT_Experiment_boxrow_deviation.pdf", bbox_inches="tight")
plt.show()
```

```python
fig, ax = plot_experiment_lines_compare(
    df_values=df_shift_dev,
    df_meta=df_experiments,
    value_col="Deviation", value_as="result. Deviation",
    x_col="Experiment_Type", x_col_as= r"Prioritätsregel",
    compare_col="Max Bottleneck Utilization",
    compare_col_as=r"$u_{\mathrm{b}}^{\max}$",
    agg_method="median",
    show_quantile_band=True,
    quantile_band=(0.25, 0.75),
    ratio_label_on=False
)
plt.show()
```

### Kendall Tau

```python
from src.utils.DataFrameEvaluation import get_kendall_tau_experiment_shift_df

df_schedules = pd.read_csv("data/schedules_v01_gt.csv")

df_kendall = get_kendall_tau_experiment_shift_df(
    df_schedules
)
df_kendall
```

```python
fig, axes = plot_experiment_boxrow(
    df_values=df_kendall,
    df_meta=df_experiments,
    value_col="Kendall_Tau",
    x_col="Experiment_Type", x_col_as= r"Prioritätsregel",
    value_as="result. Kendalls Tau",
    median_fmt = ".2f",
    ratio_label_on=False
)

fig.savefig("output/Evaluation_GT_Experiment_boxrow_kendall.pdf", bbox_inches="tight")
plt.show()
```

```python


# 🔹 Merge Werte + Meta-Daten
dfm = df_kendall.merge(
    df_experiments[["Experiment_ID", "Experiment_Type"]],
    on="Experiment_ID",
    how="left"
)

# 🔹 Gruppieren und Kennzahlen berechnen
df_summary = (
    dfm.groupby("Experiment_Type")["Kendall_Tau"]
       .agg(Mean="mean", Q25=lambda x: x.quantile(0.25), Q75=lambda x: x.quantile(0.75))
       .reset_index()
)

# 🔹 Spalten anordnen und runden
df_summary = (
    df_summary[["Experiment_Type", "Mean", "Q25", "Q75"]]
    .round(2)  # alles auf eine Nachkommastelle runden
)

df_summary
```
